```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets
match_info_data = pd.read_csv('matches.csv')
match_data_df = pd.read_csv('deliveries.csv')
```

```python

# Display dataset structure
print("Match Info Data:")
print(match_info_data.head())
print("\nMatch Data:")
print(match_data_df.head())
```

```python
# Check for missing values
print("\nMissing values in match_info_data:")
print(match_info_data.isnull().sum())
print("\nMissing values in match_data_df:")
print(match_data_df.isnull().sum())
```

```python
# Matches per city
plt.figure(figsize=(12, 6))
sns.countplot(data=match_info_data, x='city', palette='viridis', order=match_info_data['city'].value_counts().index)
plt.title("Matches per City in BBL")
plt.xticks(rotation=45)
plt.xlabel("City")
plt.ylabel("Match Count")
plt.show()
```

```python

# Teams with most wins
plt.figure(figsize=(12, 6))
winners = match_info_data['winner'].value_counts()
sns.barplot(x=winners.index, y=winners.values, palette='coolwarm')
plt.title("Most Successful Teams in BBL")
plt.xticks(rotation=45)
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.show()
```

```python
# Result margin distribution
plt.figure(figsize=(12, 6))
sns.histplot(match_info_data['result_margin'], bins=20, kde=True, color='blue')
plt.title("Result Margin Distribution")
plt.xlabel("Margin")
plt.ylabel("Frequency")
plt.show()
```

```python
# Toss Decision Analysis
plt.figure(figsize=(8, 6))
sns.countplot(data=match_info_data, x='toss_decision', palette='pastel')
plt.title("Toss Decision Trends in BBL")
plt.xlabel("Toss Decision")
plt.ylabel("Count")
plt.show()
```

```python
# Run Scoring Trends Over the Years
plt.figure(figsize=(12, 6))
sns.boxplot(data=match_data_df, x='over', y='total_runs', palette='magma')
plt.title("Run Scoring Trends Across Overs in BBL")
plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.show()
```

```python
# Matches per Venue
plt.figure(figsize=(12, 6))
sns.countplot(data=match_info_data, x='venue', palette='magma', order=match_info_data['venue'].value_counts().index[:10])
plt.title("Top 10 Venues with Most Matches")
plt.xticks(rotation=90)
plt.xlabel("Venue")
plt.ylabel("Match Count")
plt.show()
```
